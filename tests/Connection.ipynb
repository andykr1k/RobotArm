```python
import requests
from PIL import Image
from io import BytesIO
from IPython.display import display
import numpy as np
```

```python
response = requests.get('http://100.69.34.11:5000/pic_feed_OD')
```

```python
if response.status_code == 200:
    image = Image.open(BytesIO(response.content))
    display(image)
else:
    print("Failed to retrieve the image:", response.status_code)
```

```python
response = requests.get('http://100.69.34.11:5000/info')
```

```python
if response.status_code == 200:
    data = response.json()
    print(data)
    print("Distance:", data.get("distance"))
    print("Status:", data.get("on_blue"))
else:
    print("Failed to retrieve data:", response.status_code)
```
